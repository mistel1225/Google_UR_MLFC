```python
from tqdm import tqdm
```

```python
with open('data/raw_data/data35000.json', newline='') as jsonfile:
    data = json.load(jsonfile)
    for i in tqdm(range(35618)):
        with open('data/raw_data/datasets/data_{}.json'.format(i), 'w', encoding='utf-8') as f:
            json.dump(data[str(i)], f, ensure_ascii=False, indent=4)
```
